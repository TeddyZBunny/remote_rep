    Инкапсуляция
Все данные объекта должны храниться в объекте. Никто не может изменять его данные без его ведома.
    Наследование
Дочерние классы наследуют родительскую функциональность, расширяя ее и дополняя.
    Полиморфизм
Возможность перегрузки или переопределения митода


```python
from abc import abstractmethod #Для пометок абтсрактных методов

class Group:
    pupils = True
    school_num = 2
    director = 'Legenda'
    
    def __init__( self, title, count ):
        self.title = title
        self.count = count
    
    def study(self): # В методах нужно указывать self
        print('Sitting, writing, thinking and remembering')
    
    @abstractmethod    
    def move(self): # Можем пасить в методах, чтобы будущим классам оставить подсказу, что нужно будет реалзиовывать этот метод
        pass
    
class Primary_school(Group):
    max_age = 11
    min_age = 6
    floor = '2 floor'
    
    def move(self):
        # super - это значит БУДЕТ использоваться метод РОДИТЕЛЯ
        # return super().move() 
        print('Go carefully')
    
class High_school(Group):
    
    max_age = 18
    min_age =11
    floor = 'All floors'
    allowed_directions = ['math','biology', 'history', 'for_lan']
    
    # Наследование + дополнение
    def __init__(self, title, count, direction):
        super().__init__(title, count)
        if direction not in self.allowed_directions:
            raise ValueError(f"Параметр должен быть одним из: {self.direction_allowed_values}")
        self.direction = direction
    
    def move(self): # Полиморфизм - переопределили метод родительский
        print('Go slowly')

a1 = Primary_school('A1',27) 
a7 = High_school('7A',25,'math')  
print(a1.count)  

# я могу написать a1.count = 1000 - это противоречит принципу ООП и доступ к такому должен быть запрещен
```
