```python
from abc import abstractmethod #Для пометок абтсрактных методов

class Group:
    pupils = True
    school_num = 2
    director = 'Legenda'
    
    def __init__( self, title, count ):
        self.title = title
        self.count = count
    
    def study(self): # В методах нужно указывать self
        print('Sitting, writing, thinking and remembering')
    
    @abstractmethod    
    def move(self): # Можем пасить в методах, чтобы будущим классам оставить подсказу, что нужно будет реалзиовывать этот метод
        pass
    
class Primary_school(Group):
    max_age = 11
    min_age = 6
    floor = '2 floor'
    
    def move(self):
        # super - это значит БУДЕТ использоваться метод РОДИТЕЛЯ
        # return super().move() 
        print('Go carefully')
    
class High_school(Group):
    max_age = 18
    min_age =11
    floor = 'All floors'
    allowed_directions = ['math','biology', 'history', 'for_lan']
    
    def __init__(self, title, count, direction):
        super().__init__(title, count)
        if direction not in self.allowed_directions:
            raise ValueError(f"Параметр должен быть одним из: {self.direction_allowed_values}")
        self.direction = direction
    
    def move(self):
        print('Go slowly')
    
    

a1_class=Primary_school('1A', 25)
b6_class=High_school('6B', 24,'for_lan')

print(a1_class.school_num)
print(a1_class.floor)
print(a1_class.director)
print(a1_class.max_age)
a1_class.study()

print()
a1_class.move()
a1_class.study()
print(a1_class.count)

print()
b6_class.move()
b6_class.study()
print(b6_class.count)
print(b6_class.direction)

```
