```python
listik=[1,3,234,2,42,2]
new_list=[]

for x in listik:
    new_list.append(x*2)

print(new_list)
print()


new_list1=map(lambda a:a*2, listik)
print(list(new_list1))
print()

new_list2=[x*2 for x in listik]
print(new_list2)
```

```python
listik=[1,3,234,2,42,2]
new_list=[x for x in listik if x%2==0]
print(new_list)
print()

new_list2=[x if x%2==0 else x+1 for x in listik]
print(new_list2)
```

```python
listik=[1,3,234,17,42,6,23,23,57,67]

mydict={x: x+2 for x in listik}
print(mydict)
```

```python
listik = [('first', 'second'),('thirds', 'fourth')]
new_dict=dict(listik)
print(new_dict)
```

```python
countries=['USA','China', 'Canada']
temps=[25,21,10]

dict_of_temperatues = dict(zip(countries,temps))

print(dict_of_temperatues)
```
